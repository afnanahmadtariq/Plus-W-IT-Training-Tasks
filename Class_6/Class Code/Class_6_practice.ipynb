```python
#create dataframe from a dictionary

import pandas as pd
import numpy as np
employees = {
    'Emp_ID': [101, 102, 103, 104, 105],
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Emma'],
    'Age': [25, 30, 28, 35, 27],
    'Salary': [50000, 60000, 55000, 65000, 52000],
    'Department': ['HR', 'IT', 'Finance', 'Marketing', 'Operations'],
    'Location': ['New York', 'San Francisco', 'Los Angeles', 'Chicago', 'Houston'],
}

df = pd.DataFrame(employees)
print(df)
```

```python
#Handling Missing Data

df['Age'].fillna(df['Age'].mean(), inplace=True)
df['Salary'].fillna(df['Salary'].mean(), inplace=True)

df.dropna(subset=['Salary'], inplace=True)

print(df)
```

```python
#Filter Employees

it_employees = df[df['Department'] == "IT"]

fi_employees = df[df['Department'] == "Finance"]

print(it_employees)
print(fi_employees)
```

```python
#Sorting employees by Salary

df['Experience'] = np.where(df['Salary'] > 70000, 'Senior', 'Junior')

sorted_df = df.sort_values(by='Salary', ascending=False)

print(sorted_df)
```

```python
#Creating another dataframe for merging

new_hires = {
    'Emp_ID': [106, 107],
    'Name': ['Frank', 'John'],
    'Age': [44, 30],
    'Salary': [55000, 65000],
    'Department': ['HR', 'Finance'],
    'Location': ['France', 'Germany'],
}

df_newhires = pd.DataFrame(new_hires)
print(df)
```

```python
df_final = pd.concat([sorted_df, df_newhires], ignore_index=True)
print(df_final)
```

```python
df_final['Experience'] = np.where(df_final['Salary'] > 55000, 'Senior', 'Junior')

print(df_final)
```
